```python
#### You code here
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.style.use('seaborn-whitegrid')
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/unt-iialab/info5502-spring2022/main/datasets/births.csv")
data.head()

quartiles = np.percentile(data['births'], [25, 50, 75])
mu, sig = quartiles[1], 0.74 * (quartiles[2] - quartiles[0])
births = data.query('(births > @mu - 5 * @sig) & (births < @mu + 5 * @sig)')

births['day'] = births['day'].astype(int)

births.index = pd.to_datetime(10000 * births.year + 100 * births.month + births.day, format='%Y%m%d')
births_by_date = births.pivot_table('births', [births.index.month, births.index.day])
births_by_date.index = [pd.datetime(2012, month, day) for (month, day) in births_by_date.index]

fig, ax = plt.subplots(figsize=(12, 6))
plt.title("USA births by day of a year from 1969 to 1988")
plt.xlabel("Months of a year")
births_by_date.plot(ax=ax);

```

<span style='background:yellow'>**Question 4 (60 pt, each sub-question is 20 pt).**</span> In the world of business, the importance of the customers reviews for any product is not ignorable. The impact of customers reviews on customers decision making is significant. Therefore, businesses try to increase this impact by increasing the ratio of the positive reviews to negative reviews. One effective way to increase the ratio is offering the incentives such as small gifts or free next services to customers to post the reviews for item they purchase or used. However, it is not always the case, and customer may post their reviews without receiving or accepting the incentives. So, the reviews could be in either incentivized or non-incentivized category.


Considering the importance of reviews, in this question you need to compare these two categories of reviews. One method to compare the effectiveness of two categories is conducting the A/B testing.


The Software_Reviews dataset is collected from software review products, and it is combination of several small datasets. Find the dataset as an attachment on the left side of the page.


You can downlaod the dataset from github: https://github.com/unt-iialab/info5502-spring2022/blob/main/finalexam/Software_Reviews_csv_files_S.csv

(1) Clean the dataset before you use (You need to know this dataset is combinations of the several CSV files with title rows. So, there are extra title rows in the middle of the file, which they need to be removed.)


(2) Apply an A/B test": consider the incentivized column as main column for A & B grps.

a. The incentivized column includes NominalGift, VendorReferredIncentivized, NoIncentive, NonNominalGift, VendorReferred, None

b. grp A (Incentive): NominalGift, and VendorReferredIncentivized

c. grp B (NoIncentive): NoIncentive, NonNominalGift, VendorReferred

d. Make the decision if None includes in either grp A or B or remove that.

e. The OverAllRating is another column that is used to compare grps A and B.

Which one A or B grp has higher OverAllRating?


(3) Draw the A/B testing related visualizations.

```python
#### You code and your answer here here
from google.colab import drive 
#drive.mount('/content/gdrive')

data = pd.read_csv("/content/Software_Reviews_csv_files_S.csv")

import numpy as np
from scipy.stats import norm

import matplotlib.pyplot as plt

#.shape
grp_a = data[(data['incentivized'] == "NominalGift") | (data['incentivized'] == "VendorReferredIncentivized")]['OverAllRating']
grp_b = data[(data['incentivized'] == "NoIncentive") | (data['incentivized'] == "NonNominalGift") | (data['incentivized'] == "VendorReferred")]['OverAllRating']#.shape

N_con = grp_a.shape[0]
N_exp = grp_b.shape[0]

# Significance Level
alpha = 0.05


mu_con = np.mean(grp_a)
mu_exp = np.mean(grp_b)

z = grp_a.astype('float64')
variance_con = np.var(z)

k = grp_b.astype('float64')
variance_exp = np.var(k)

# Pooled Variance
pld_variance = np.sqrt(variance_con/N_con + variance_exp/N_exp)

# Test statistics
Te = (mu_con-mu_exp)/np.sqrt(variance_con/N_con + variance_exp/N_exp)

# two sided test and using symmetry property of Normal distibution so we multiple with 2
p_value = norm.sf(Te)*2

# Z-critical value
Z_crit  = norm.ppf(1-alpha/2)

# Margin of error
Margin = Z_crit*pld_variance

# Confidence Interval
Confidence = [(mu_con - mu_exp) - Margin, (mu_con - mu_exp) + Margin]


print("Test Statistics stat: ", Te)
print("Z-critical: ", Z_crit)
print("P_value: ", p_value)
print("Confidence Interval of two sample Z-test for proportions: ", np.round(Confidence,2))

import matplotlib.pyplot as plt
z = np.arange(-3,3,  0.1)
plt.plot(z, norm.pdf(z), label = 'Standard Normal Distribution',color = 'orange',linewidth = 3.5)
plt.fill_between(z[z>Z_crit], norm.pdf(z[z>Z_crit]), label = 'Rejection Region on Right',color ='b' )
plt.fill_between(z[z<(-1)*Z_crit], norm.pdf(z[z<(-1)*Z_crit]), label = 'Rejection Region on Left',color ='b' )
plt.title("Samples of two Z-test rejection region")
plt.legend()
plt.show()

```
